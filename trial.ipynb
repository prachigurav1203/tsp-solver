```python
from functools import lru_cache

def tsp(dist):
   
    n = len(dist)
    ALL_VISITED = (1 << n) - 1

    @lru_cache(None)
    def dp(mask, pos):
        # If all cities visited, return cost to go back to start
        if mask == ALL_VISITED:
            return dist[pos][0]

        ans = float('inf')
        for city in range(n):
            if not mask & (1 << city):
                new_cost = dist[pos][city] + dp(mask | (1 << city), city)
                ans = min(ans, new_cost)
        return ans

    return dp(1, 0)  # start at city 0


# Example usage:
dist_matrix = [
    [0, 29, 20, 21],
    [29, 0, 15, 17],
    [20, 15, 0, 28],
    [21, 17, 28, 0]
]

print("Minimum TSP tour cost:", tsp(dist_matrix))
```
